```python
# identifierHash =익명의 고유ID
# type =엔티티의 유형 (모두 사용자 값임)
# country (국가) = 사용자 국가 (프랑스어로 작성) , ISO 식별자를 선호하는 경우는 "국가코드"를 참조하세요
# language (언어)= 사용자가 선호하는 언어(사이트 사용시 인터페이스 언어)
# socialNbFollowers (팔로워) = 이 사용자의 활동을 팔로우하는 사용자 수. 새로운 계정은 자동으로 매장의 공식 계정으로 이어집니다.
# socialNbFollows (팔로우) = 이 사용자가 팔로우하는 사용자 계정의 수. 공식 파트너를 따르도록 새 계정이 자동으로 할당됩니다.
# socialProductsLiked (제품 좋아요) = 이 사용자가 좋아하는 제품 수
# productsListed (제품 목록) = 이 사용자가 업로드 한 현재 판매되지 않은 제품의 수
# productsSold (판매 제품 수) = 이 사용자가 판매 한 제품 수
# productsPassRate (제품 통과율) = 제품 설명을 충족하는 제품의 %. (판매 된 제품은 구매자에게 배송되기 전에 상점 팀이 검토합니다.)
# productsWished (원하는 제품) = 이 사용자가 자신의 위시리스트에 추가 한 제품 수.
# productsBought (구입 제품) = 이 사용자가 구입 한 제품 수
# gender (성별) = 사용자의 성별
# civilityGenderId (성별 정수화)  = 정수로서의 시민
# civilityTitle (성별 문자화) = 시민 제목       
# hasAnyApp (공식앱 이용 여부) = 사용자는 매장의 공식 앱을 사용한 적이 있습니다.         
# hasAndroidApp (Android 앱 이용 여부) = 사용자는 공식 Android 앱을 사용한 적이 있습니다         
# hasIosApp (아이폰앱 이용여부)= 사용자는 공식 iOS 앱을 사용한 적이 있습니다       
# hasProfilePicture(프로필 유무) = 사용자는 사용자 정의 프로필 사진이 있습니다    
# daysSinceLastLogin (마지막 로그인 이후 일 수)
# seniority (사용자 등록 후 경과 일)       
# seniorityAsMonths( 개월 ) = 선배를 참조하십시오. 몇 달 안에        
# seniorityAsYears( 연도 ) =선배를 참조하십시오. 몇년에 걸쳐        
# countryCode (국가 코드) = 사용자 국가 (ISO-3166-1)
```

```python
# 컬럼간의 상관관계 분석

# 데이터프레임의 상관관계 계산
corr = df1.corr()

# 플롯의 크기를 설정
plt.figure(figsize=(12,10))

# 히트맵 생성
# annot=True 상관계수를 표시, fmt=".1f"소수점 한 자리 표시.
# cmap='coolwarm'은 색상 옵션
# figsize, plt.figure()를 메소드로 크기 조정.
sns.heatmap(corr, annot=True, fmt=".1f", cmap='coolwarm')

# 축 라벨이 겹치지 않게 회전
plt.xticks(rotation=90)
plt.yticks(rotation=0)

# 플롯을 보여주기
plt.show()
```

```python
df2 = df1.select_dtypes(include=['int64', 'float64']) # 수치형 데이터값이 있는 컬럼만 가져오기
```

```python
## 저분산인 feature 제거

from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold()
transformed_df = selector.fit_transform(df2)  # numpy.ndarray 형태. 
df2 = pd.DataFrame(transformed_df, columns=selector.get_feature_names_out())
df2
```

```python
corr2 = df2.corr()
```

```python
# 플롯의 크기를 설정
plt.figure(figsize=(9,7))


# annot은 수치 표기, fmt은 숫자 자릿수 표기, coolwarm은 색상
sns.heatmap(corr2, annot=True, fmt='.1f', cmap='coolwarm')
```

```python
# 만약 i 400 이하이면, 'Customer_Status'를 'maintain'(유지 고객)설정.
# 만약 i가 400 초과이면, 'Customer_Status'를 'leave'(이탈 고객)로 설정. 

# 조건에 따라 'Customer_Status' 컬럼 생성
df2['Customer_Status'] = ['maintain' if i <= 400 else 'leave' for i in df1['daysSinceLastLogin']]
df2
```

```python
# countplot 함수는 카테고리별 데이터 개수를 세어서 막대 그래프로 표시.
# x 값은 막대 그래프의 x축을 df2 데이터프레임의 civilityGenderId 컬럼으로 설정.
# hue 값은 색상을 구분, 각 civilityGenderId 값의 분포를 다른 색상의 막대로 표시.
# data 값은 사용할 데이터프레임을 df2로 지정.

# 유지 고객과 이탈 고객의 성별 분포
sns.countplot(x='civilityGenderId', hue='Customer_Status', data=df2)
plt.show()
```
